```python
from docx import Document
from docx.shared import Inches
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
FIGDIR = 'namespace'
Path(FIGDIR).mkdir(parents=True, exist_ok=True)
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.savefig(FIGDIR + '/sample_figure.png')
```

```python
# template file
doc = Document('template.docx')
tables = doc.tables

# first tables 
p = tables[0].rows[1].cells[0].add_paragraph()
r = p.add_run()
# figure go there
r.add_picture(FIGDIR + '/sample_figure.png')

# second tables 
p = tables[1].rows[1].cells[0].add_paragraph()
r = p.add_run()
r.add_picture(FIGDIR + '/sample_figure.png',width=Inches(4.0))
doc.save('bootstrap.docx')
```
